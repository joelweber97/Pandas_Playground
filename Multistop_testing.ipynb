```python
import pandas as pd
import numpy as np
```

```python
shot_num = np.random.randint(0,200,5000000).tolist()
shot_num[:5]
```

```python
chan = np.random.randint(1,101, 5000000).tolist()
chan[:5]
```

```python
x = np.random.uniform(0.0,50, 5000000).tolist()
x[:5]
```

```python
df = pd.DataFrame(list(zip(shot_num,chan,x)), columns = ['shot_num','chan','x'])
df.head()
```

```python
gb = df.groupby(['shot_num','chan'])
```

```python
#convert each groupby into a list
gb = gb.aggregate(lambda x: x.tolist())
gb.head(2)
```

```python
#sort the groupby lists in ascending order 
gb = gb['x'].apply(sorted)
gb.head(2)
```

```python
#subtract i[n] from i[n+1] for every list in gb['x']
def diff(l):
    nl = [(n-l[i-1]) if i else n for i,n in enumerate(l)]
    return nl[1:]
gb = gb.apply(diff)
gb.head(2)
```

```python
def append_to_list(l2):
    final_list.append(l2)
    return final_list
final_list = []
gb.apply(append_to_list)
```

```python
flat_list = [j for i in final_list for j in i]
```
