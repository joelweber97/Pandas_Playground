### Create a MultiIndex on a DataFrame with the .set_index() method

```python
mac = pd.read_csv('./data/bigmac.csv', parse_dates = ['Date'])
mac.head(3)
```

```python
mac.set_index(keys = ['Date', 'Country'])
#can input a list into the .set_index() method to create a multiindex df
#the date is the outer level index
#the country is the internal layer index
#the order of the list input determines the order of the order of the keys in the multiindex
```

```python
#most outer index should be the column with the least number of unique values.
mac.nunique()
#can se that we have 12 unique dates and 58 unique countries and 330 unique prices
```

```python
mac.set_index(keys = ['Date', 'Country'], inplace = True)
mac.head(3)
```

```python
#can also sort these index's like any normal df
mac.sort_index(inplace = True)
#this will actually sort ALL the index's. It'll start by sorting the outermost level and working inward
#the earliest date will be first and the last date will be last. Within each date the countries will be sorted alphabetically.
```

```python
mac.head(3)
```

```python
mac.index
#returns a multiindex object. that's really just a list of tuples consisting of the combinations of index's.
#to find a value of a cell 1 spcific index label isn't enough. We can't search for a date and get a specific cell back.
#we need to also put input the country label to find that specific cell we're looking for.
```

```python
mac.index.names
```

```python
#another way to create a multiindex on the read
mac = pd.read_csv('./data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date','Country'])
mac.head()
```

### Change the Index Level Name with the .set_names() method

```python
mac = pd.read_csv('./data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date','Country'])
mac.sort_index(inplace = True)
mac.head(3)
```

```python
#set_names() is also called on the index of the df
mac.index
```

```python
mac.index.set_names(['Day', 'Location'],inplace = True)
#the first parameter is the names.
#can pass in a list of new names if you want to change all the names
mac.head()
```

```python
#if you only want change one you can still pass in a list and keep the same name in that list.
#mac.index.set_names(['Day', 'Country'],inplace=True)
#another way of doing this is by using the level parameter
#can pass in the index position or the name of the current index name
mac.index.set_names('Date', level=0)
mac.index.set_names('Date', level='Day',inplace = True)
mac.index
```

### The .sort_index() method on a MultiIndex DataFrame

```python
mac = pd.read_csv('./data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date','Country'])
mac.head(3)
```

```python
mac.sort_index()
#sorts the outermost level in ascending order (oldest to newest)
#then sorts the inner level in ascending order (alphabetically a to z)
#since the default argument for the ascending parameter is True
```

```python
#can also pass false in the ascnding parameter
mac.sort_index(ascending = False)
```

```python
#can customize the sort order for each level
mac.sort_index(ascending = [True, False])
mac.sort_index(ascending = [False, True], inplace=True)
mac.head()
```

```python
#can also sort a single index using the level parameter
mac.sort_index(level = 0)
mac.sort_index(level = "Date")
#both sort the date index ascending but leave the country index in the same order it was in
```

```python
mac.sort_index(level=1)
mac.sort_index(level='Country')
#this sorts only by country.
```

### Extract Rows from a MultiIndex DataFrame using the .loc[] and .iloc[] accessor

```python
mac = pd.read_csv('./data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date','Country'])
mac.sort_index(inplace = True)
mac.head(3)
```

```python
#first we cover an alternate method of doing this that isn't really preferred.
mac.loc['2010-01-01','Argentina']
#usually when using .loc[] the first argument refers to the index and the second argument refers to the column
#here the first argument refers to the first index and the second argument refers to the second column.
#so this syntax doesn't reallly match up with the standard pandas way of using .loc
```

```python
#better option
#pass a tuple in with the index labels as the first argument and the second argument is the column/s name you want to target in a list
mac.loc[('2010-01-01', 'Argentina'), 'Price in US Dollars']
```

```python
#multiIndex's don't affect iloc at all
#each row is still given an index position in line
#row 1 will be the first row in line.
#doesn't start counting within each level, starts counting from the top.
mac.iloc[0]
mac.iloc[500]
```

```python
mac.iloc[200:500:50]
```

```python
mac.iloc[[10,30,40,300,301,500]]
```

### The .transpose() method on a MultiIndex DataFrame

```python
mac = pd.read_csv('./data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date','Country'])
mac.sort_index(inplace = True)
mac.head(3)
```

```python
#swaps the index's of the df.
#row labels are now the column headers
#takes the column headers and move them to the row labels.
mac.transpose()
#actually creates multiindex columns
```

```python
#tranpose doesn't mutate a df nor does it have an inplace parameter so we have to assign it back to a varaible
mac = mac.transpose()
mac.head()
```

```python
#slicing using the .loc[] accessor after transpose
mac.loc[("Price in US Dollars",),('2010-01-01',)]
```

```python
mac.loc[('Price in US Dollars',), ('2010-01-01','Sri Lanka')]
```

```python
mac.loc[('Price in US Dollars',),('2010-01-01',)].values
```

```python
#.loc[] slicing also works on multiIndex df
#pull out values of sri lanka through ukraine for jan 1 2010
mac.loc[('Price in US Dollars',),('2010-01-01','Sri Lanka'):('2010-01-01','Ukraine')]
```

### The .unstack() method, Part 1

```python
world = pd.read_csv('./data/worldstats.csv', index_col = ['country','year'])
world.head(3)
```

```python
#gonna call the stack method first to create the 3 level index we saw in the lesson above
s = world.stack()
```

```python
s.head()
```

```python
s.unstack()
#innermost layer (population and gdp) is moved back into the column names
```

```python
s.unstack().unstack() 
#innermost 2 layers (population/gdp and year) are moved to the columns
```

```python
s.unstack().unstack().unstack()
#all 3 index levels have been moved into the columns.
```

### The .unstack() method, Part 2

```python
world = pd.read_csv('./data/worldstats.csv', index_col = ['country','year'])
world.head(3)
s =world.stack()
s.head()
```

```python
#various parameters that can be added to the method to specify which index level will be added to the columns
#s.unstack(level = -1, fill_value = None)
#the level is index location of the index level
#outer most index (Country) has a index location of 0
#next level (year) has an index location of 1.
#level = -1 is the inner most index. Similar to the -1 index of a list being the last value.
```

```python
#want to move year into the columns
s.unstack(level = 1)
#moves year across the top and country and pop/gdp down the side
```

```python
#this sets the countries across the top and has a multiIndex of the year and pop/gdp
s.unstack(level = 0)
```

```python
# can also use -n for the level
s.unstack(level = -1) #which is the same as s.unstack(2)
```

```python
s.unstack(level = -2)
#this is the same as level = 1
```

```python
#can also write the string name for the index
s.unstack('country')
```

```python
s.unstack('year')
```

### The .unstack() method, Part 3

```python
world = pd.read_csv('./data/worldstats.csv', index_col = ['country','year'])
s = world.stack()
s.head(3)
```

```python
#can unstack by multiple levels to columns axis in 1 move.
#can pass in a list with the index positions or the index name to unstack those levels
s.unstack(level = [0,1])
s.unstack(level = ['country','year'])
#order matters. First number in list will be outer most column header
```

```python
s.unstack(level = [1,0])
s.unstack(level = ['year','country'])
```

```python
s.unstack([2,0])
```

```python
s = s.unstack('year', fill_value = 0)
s
```

### The .pivot() method

```python
#.pivot() is used to re-orient a dataset
#it takes values in a column and converts them to column headers
```

```python
sales = pd.read_csv('./data/salesmen.csv',parse_dates= ['Date'])
#of the 1830 rows there are only 5 salesman so we'll convert that to category
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head()
```

```python
#we want to take the 5 different salesman that occur i nthe salesman column and move that to column headers.
sales = sales.pivot(index = 'Date', columns = 'Salesman', values = 'Revenue')
sales.head(3)
#index is the column we want to be the index after the pivot
#columns is the column values we want to take and spread across the top
#values are what we want to input into the cells under each saleman column.
#not usefull if the columns parameter has a lot of different values in that original column.
```

```python
#makes it easier to do math on a specific salesman or across dates. We can see which date had the most sale
sales.sum(axis = 1).max(axis = 0)
```

### Use the .pivot_table() method to create an aggregate summary of a DataFrame

```python
#works exactly the same as it does in xcel
```

```python
foods = pd.read_csv('./data/foods.csv')
foods[['Gender','City','Frequency','Item']] = foods[['Gender','City','Frequency','Item']].astype('category')
foods.head(3)
```

```python
#average spent per gender
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'mean')
```

```python
#total sales per city
foods.pivot_table(values = 'Spend', index='City', aggfunc = 'sum')
```

```python
#total sales by item
foods.pivot_table(values = 'Spend',index='Item',aggfunc = 'sum').sort_values('Spend')
```

```python
#how much each gender bought of each item.
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], aggfunc = 'sum')
#as you can see, passing in a list as the index parameter returns a multiIndex series
```

```python
#we can also add more columns to make it into a df
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = 'City', aggfunc = 'mean')
```

```python
#can also have a multiindex columns
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['Frequency','City'],aggfunc = 'mean')
```

```python
#total spend on each item in each city for each gender
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['City'],aggfunc = 'sum')
#average spent on each item in each city for each gender
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['City'],aggfunc = 'mean')
#number of that item in each city by gender
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['City'],aggfunc = 'count')
#the max in the spend column for each item in each city for each gender
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['City'],aggfunc = 'max')
#the min in the spend column for each item in each city for each gender
foods.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['City'],aggfunc = 'min')
```

### Use the pd.melt() method to create a narrow dataset from a wide one

```python
#works as the reverse operation of a pivot() method
#takes an aggregated dataset and turns it into tabular format.
sales = pd.read_csv('./data/quarters.csv')
sales
```

```python
pd.melt(sales, id_vars = 'Salesman')
#new 3 column dataset that keeps the salesman and adds a variable column (value that was initially stored as column header)
#and the value associated with that salesman/quarter combination.
#condensed the column headers into a single row and added necessary salesman to match 
```

```python
#we can change the column names so they're not variable and value in the melted dataset
test = pd.melt(sales, id_vars = 'Salesman', var_name = 'Quarter', value_name = 'Revenue')
```

```python
test
```

```python
#reverse of the pd.melt method is the .pivot() method
test.pivot(index = 'Salesman', columns = 'Quarter',values = 'Revenue')
```
