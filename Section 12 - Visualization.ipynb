### Creating Bar Graphs to Show Counts

```python
bb = data.DataReader('BB',data_source='yahoo',start='2007-01-01',end='2020-12-31')
bb.head(3)
```

```python
#lets focus on close column
#categorize this column into 3 groups (poor, satisfactory, great)
bb['Close']
```

```python
#we can use the .apply method with a custom function to classify the close column
def performance(stock_price):
    if stock_price < 10:
        return 'Poor'
    elif stock_price <=50:
        return 'Satisfactory'
    else:
        return 'Great'
```

```python
bb['Close'].apply(performance)
#returns a new series based on the classifcation of each date
```

```python
bb['Close'].apply(performance).value_counts()
#returns another new series with the performance ranks as index and the count of each as the value
```

```python
bb['Close'].apply(performance).value_counts().plot(kind = 'bar')
#this will actually create the bar graph using the kind parameter
```

```python
#can also do a horizontal bar graph
bb['Close'].apply(performance).value_counts().plot(kind = 'barh')
```

### Creating Pie Charts to Represent Proportions/Ratios

```python
bb = data.DataReader('BB',data_source='yahoo',start='2007-01-01',end='2020-12-31')
bb.head(3)
```

```python
#want to find how many of our stock performance days come in below the average and above the average
#start by calculating the mean
bb['Close'].mean()
```

```python
#then create another function that can be applied to every cell that creates a higher or lower rank
def performance(stock_price):
    if stock_price > bb['Close'].mean():
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
#then run the performance function on the close column using the .apply() method
bb['Close'].apply(performance)
```

```python
#then do the value_counts
bb['Close'].apply(performance).value_counts()
```

```python
plt.style.use('seaborn-poster')
bb['Close'].apply(performance).value_counts().plot(kind = 'pie')
#tells you there are more below average days than above average days
#which means there are some above average days that are so high they drag the average up
```
