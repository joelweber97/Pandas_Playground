### Drop DataFrame Rows with Null Values with the .dropna() method

```python
nba = pd.read_csv('./data/nba.csv')
nba
```

```python
#dropna removes any dataframe row that has any Nan Values in it.
#can also drop any column with any Nan values in it by changing axis = 1
#can also use the inplace= True argument to run the dropna inplace
#by default the how parameter is set to any, so if any column in that row has Nan it will be dropped
nba.dropna()
```

```python
#if we change the how parameter to 'all', the row will be dropped if all the columns in that row are NaN
nba.dropna(how = 'all', inplace = True)
#only the last row was removed because every column in that row was NaN.
#any row with only 1 NaN will still be there.
#we also set inplace = True to permenantly remove those rows.
```

```python
#now we have a df with rows with all NaN values removed
#remove any columns with any NaN values
nba.dropna(axis = 1)
```

```python
#how to remove a row if there's a null value in the salary column, but we don't care if there's null values
#in any other column
nba.dropna(subset = ['Salary'])
#removes any row that contains a NaN in the salary column. There can be NaN in other columns and that row won't drop.
```

```python
nba.dropna(subset = ['College'])
#this now removes any row where there was an NaN in the college column but not in the salary column.
```

```python
nba.dropna(subset = ['College', 'Salary'])
#now we drop any row where the college column and salary column contain NaN
```

### Fill in Null DataFrame Values with the .fillna() method

```python
nba = pd.read_csv('./data/nba.csv')
nba.head(3)
```

```python
#nba.fillna(value = None, method = None, axis = None, inplace = False, limit = None, downcast = None)
nba.fillna(0)
#we have the NaN values in salary replaced with 0, which is good.
#we also have NaN values in the college column replaced with a 0, which is not good.
```

```python
#makes sense to call the fillna on a series from the df rather than the df as a whole.
nba['Salary'].fillna(0)
#but we stil have to put that series back into the df using the inplace = True argument.
```

```python
nba['Salary'].fillna(0, inplace = True)
nba
#the nba df now as all the NaN values in the salary column replaced with 0 which all NaN values in any
#other column still remain.
```

```python
nba['College'].fillna('No College', inplace = True)
nba
#now the NaN values in the college column was replace with the string 'No College'
```

### Convert DataFrame Column TYpes with the .astype() method

```python
#.astype() allows us to convert the column type of a df from one to another type.
#.astype() requires a column series not to have any null values.
nba = pd.read_csv('./data/nba.csv').dropna(how = 'all')
nba.tail(3)
```

```python
#we still have nan values in the salary and college column so we have to fill those
nba['Salary'].fillna(0, inplace = True)
nba.head()
```

```python
nba['College'].fillna('No College', inplace = True)
nba.tail(3)
#now we have a df will all the nan values dropped or filled with a different value.
```

```python
nba.dtypes
nba.info()  #we can see each column name with the number of null values in the column, and the dtype of that column.
```

```python
# the values in the initial csv that pd read was integers. When theres a null value in a particular column, pd
#will automatically read numeric values as floats instead.
#now that we have all nan values removed we will convert those numeric values back into integers.
```

```python
nba['Salary'].astype('int')  #returns a brand new series where all the values have changed to integers.
#unfortunately, astype() doesn't have an inplace parameter so we have to set it manually.
```

```python
#setting the results from teh series above back into the initial df.
nba['Salary'] = nba['Salary'].astype('int')
nba.head()
```

```python
nba.info()  # we can now see that the dtype for salary column is int32
```

```python
nba['Age'] = nba['Age'].astype('int')
nba['Number'] = nba['Number'].astype('int')
```

```python
nba.info()
```

```python
#pandas also has a unique dtype called 'Category'
#ideal when theres only a small number of unique values in a df column
#eg if you have a gender column in a df with millions of rows. Make that column a Category.
#The category dtype will reduce memory usage.
#rather than using a string to store a unique string in each cell, it points every cell 
#that has that same value to a single object
#in this example we'll set the position column to a category dtype.
#nba['Position'].nunique()  # shows that there are only 5 positions out of the 450 rows.
#might as well make that into a category so we don't have to store each string as a unique string, but rather
#we store each unique string as it's own object that other strings will point to.
nba['Position'] = nba['Position'].astype('category')
```

```python
nba.info()
```

```python
#teams would also be a great column to convert to a category dtype
nba['Team'] = nba['Team'].astype('category')
nba.info()
```

### Sort a DataFrame with the .sort_values() method, Part 1

```python
nba = pd.read_csv('./data/nba.csv')
nba.head(3)
```

```python
#when you call the .sort_values() on a series it just sorts that single column
#using the .sort_values() on a df we will need to specific which column to sort by.
#nba.sort_values(by,axis=0,ascending=True,inplace=False,kind='quicksort',na_position='last',ignore_index=False)
nba.sort_values('Name')
```

```python
nba.sort_values('Age', ascending = False)
```

```python
nba.sort_values('Salary', ascending = False, ignore_index=True)
#the ignore index will reset the index back to 0-457 after the sorting as taken place. Now Kobe will have
#the index value of 0.
```

```python
nba.sort_values('Salary', ascending = False, inplace = True)
nba
```

```python
#If the NaN values aren't removed they will be sorted to the very end.
#that's because the 'na_position' argument is set to 'last'
nba.sort_values('Salary').tail()
```

```python
#can change the nan value position by setting the na_position to 'first'
nba.sort_values('Salary', na_position='first')
```

### Rank Series Values with the .rank() method

```python
nba = pd.read_csv('./data/nba.csv')
nba.head(3)
#call this on a series to produce a brand new series of ranks
```

```python
#we have to get of those null values to get the .rank() method to work
nba.dropna(how = 'all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba['College'] = nba['College'].fillna('No College').astype('int')
nba.head()
```

```python
#we want to rank each player based on their salary
#by default the highest salary is at the end of the ranking
#and the smallest salary is being ranked at the beginning.
nba['Salary'].rank()
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending = False).astype('int')
nba.head()
#this creates a new 'Salary Rank' column in the nba df where the highest salary is the lowest rank
```

```python
#we can sort by salary to see that the rank column works correctly
nba.sort_values('Salary', ascending = False)
```
